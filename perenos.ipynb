# Численное решение

```python
import numpy as np
import matplotlib.pyplot as plt

M = 100+1
N = 100+1
epsilon = 0.00001
t_start = 0
x_start = 0
x_end = -1
t_end = 1

h = (x_end - x_start)/(N-1)
tau = (t_end - t_start)/(M-1)
y1 = np.zeros((M,N))

for n in np.arange(N):
 y1[0][n] = -h*n # initial
for m in np.arange(M):
 y1[m][0] = 0 # bounder

def F(m, n):
 return np.arctan(y1[m][n])

def der_f(m1,n1):
 return (1/(2.*tau)) - 1/(2.*h*(1 + (y1[m1][n1])*(y1[m1][n1])))

def f(mp1,np1):
  n = np1-1
  m = mp1-1
  return (y1[mp1][n] - y1[m][n] + y1[mp1][np1] - y1[m][np1])/(2.* tau)-(F(mp1, np1)-F(mp1, n) + F(m, np1) - F(m, n))/(2.* h)

for m in np.arange(M)[0:M -1]:
 for n in np.arange(N)[0:N -1]:
  eps = epsilon + 1
  while eps > epsilon :
    ep = f(m +1, n +1)/der_f(m+1,n +1)
    y1[m+1][n+1] = y1[m+1][n +1] - ep
    eps = np.abs(ep)


# Graphics
xn = np.linspace(x_start, x_end, num = N)


# 2D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
plt.ylim(-0.05, 1)
plt.xlim(-1, 0.05)

plt.ylabel('u')
plt.xlabel('x')
p = ax.scatter(xn, y1[0][:], color = "blue")
p = ax.scatter(xn, y1[25][:], color = "green")
p = ax.scatter(xn, y1[50][:], color = "cyan")
p = ax.scatter(xn, y1[75][:], color = "magenta")
p = ax.scatter(xn, y1[100][:], color = "red")
plt.grid(True)
fig.show()

```
